```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('/content/train_and_test2.csv') # Training set is already available
train.head()
```

```python
train.info(verbose=True)
```

```python
d=train.describe()
```

**Exploratory analysis and plots**

*Bar plot*

```python
dT=d.T
dT.plot.bar(y='count')
plt.title("Bar plot of the count of numeric features",fontsize=17)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='2urvived',data=train,palette='RdBu_r')
sns.pairplot(train)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='2urvived',hue='Sex',data=train,palette='RdBu_r')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='2urvived',hue='Pclass',data=train,palette='rainbow')
```

```python
f_class_survived=train.groupby('Pclass')['2urvived'].mean()
f_class_survived = pd.DataFrame(f_class_survived)
f_class_survived
f_class_survived.plot.bar(y='2urvived')
sns.countplot(x='2urvived',data=f_class_survived,palette='rainbow')
plt.title("Fraction of passengers survived by class",fontsize=17)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='2urvived',hue='SibSp',data=train,palette='rainbow')
```

```python
plt.xlabel("Age of the passengers",fontsize=18)
plt.ylabel("Count",fontsize=18)
plt.title("Age histogram of the passengers",fontsize=22)
#train['Age'].hist(bins=30,color='darkred',alpha=0.7,figsize=(10,6))
train['Age'].hist()
```

```python
plt.figure(figsize=(12, 10))
plt.xlabel("Passenger Class",fontsize=18)
plt.ylabel("Age",fontsize=18)
sns.boxplot(x='Pclass',y='Age',data=train,palette='winter')
```

```python
f_class_Age=train.groupby('Pclass')['Age'].mean()
f_class_Age = pd.DataFrame(f_class_Age)

f_class_Age.plot.bar(y='Age')
plt.title("Average age of passengers by class",fontsize=17)
plt.ylabel("Age (years)", fontsize=17)
plt.xlabel("Passenger class", fontsize=17)
```
